      3.Create new features

```python
# 1. Overall Performance Index (average of all three grades)
df['performance_index'] = (df['G1'] + df['G2'] + df['G3']) / 3
print("Performance Index (average of G1, G2, G3):")
print(df['performance_index'].describe())
print("\n" + "="*50 + "\n")
```

```python
# 2. Alcohol Consumption Score (combined workday and weekend consumption)
df['alcohol_score'] = df['Dalc'] + df['Walc']
print("Alcohol Consumption Score (Dalc + Walc):")
print(df['alcohol_score'].describe())
print("\n" + "="*50 + "\n")
```

```python
# 3. Academic Risk Indicator (based on failures and low grades)
df['academic_risk'] = ((df['failures'] >= 2) | (df['G3'] < 10)).astype(int)
print("Academic Risk Indicator (1=At Risk, 0=Not at Risk):")
print(df['academic_risk'].value_counts())
print(f"At-risk students: {df['academic_risk'].sum()} ({df['academic_risk'].mean()*100:.1f}%)")
print("\n" + "="*50 + "\n")
```

```python
# 4. Study Efficiency Ratio (performance per unit study time)
df['study_efficiency'] = df['performance_index'] / (df['studytime'] + 0.01)  # Add small constant to avoid division by zero
print("Study Efficiency Ratio (Performance per study time unit):")
print(df['study_efficiency'].describe())
print("\n" + "="*50 + "\n")
```

```python
# 5. Parental Education Index (average of mother and father education)
df['parental_edu_index'] = (df['Medu'] + df['Fedu']) / 2
print("Parental Education Index:")
print(df['parental_edu_index'].describe())
print("\n" + "="*50 + "\n")
```

     6. Outlier Detection and Handling 

```python
# PART 1: IQR Method for Outlier Detection
numerical_cols = ['age', 'Medu', 'Fedu', 'traveltime', 'studytime', 'failures', 
                  'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 
                  'absences', 'G1', 'G2', 'G3']

print("OUTLIER DETECTION USING IQR METHOD")
print("="*50)

outlier_results = []
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_results.append([col, len(outliers), f"{len(outliers)/len(df)*100:.2f}%", round(lower,2), round(upper,2), df[col].min(), df[col].max()])

outlier_df = pd.DataFrame(outlier_results, columns=['Feature', 'Outlier Count', 'Outlier %', 'Lower Bound', 'Upper Bound', 'Min', 'Max'])
print(outlier_df.to_string(index=False))
print("\n" + "="*50)
```

```python
# PART 2: Boxplot Visualization
fig, axes = plt.subplots(4, 4, figsize=(16, 12))
for i, col in enumerate(numerical_cols):
    sns.boxplot(y=df[col], ax=axes[i//4, i%4])
    axes[i//4, i%4].set_title(col)
plt.tight_layout()
plt.show()
```

```python
# PART 3: Absences - Extreme Outlier Analysis
print("\nABSENCES OUTLIER ANALYSIS")
print("="*50)
Q1_abs = df['absences'].quantile(0.25)
Q3_abs = df['absences'].quantile(0.75)
IQR_abs = Q3_abs - Q1_abs
upper_abs = Q3_abs + 1.5 * IQR_abs
print(f"Upper bound: {upper_abs:.2f}")
print(f"Max absences: {df['absences'].max()}")
print(f"Students >20 absences: {len(df[df['absences'] > 20])}")
print(f"Students >30 absences: {len(df[df['absences'] > 30])}")
print(f"Students >50 absences: {len(df[df['absences'] > 50])}")

extreme = df[df['absences'] > 30]
print(f"\nAvg G3 (all students): {df['G3'].mean():.2f}")
print(f"Avg G3 (>30 absences): {extreme['G3'].mean():.2f}")
```
